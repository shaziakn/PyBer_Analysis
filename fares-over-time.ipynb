```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
# NumPy and Statistics
# Import NumPy and the stats module from SciPy.
import numpy as np
import scipy.stats as sts
```

```python
# Files to load
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"
```

```python
city_data_df = pd.read_csv(city_data_to_load)
```

```python
ride_data_df = pd.read_csv(ride_data_to_load)
```

```python
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])
```

```python
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
```

```python
# Total rides
urban_ride_count = urban_cities_df.groupby(["city"]).count()["ride_id"]
total_rides_urban = urban_ride_count.sum()

suburban_ride_count = suburban_cities_df.groupby(["city"]).count()["ride_id"]
total_rides_suburban = suburban_ride_count.sum()

rural_ride_count = rural_cities_df.groupby(["city"]).count()["ride_id"]
total_rides_rural = rural_ride_count.sum()
```

```python
# Total drivers
urban_driver_count = urban_cities_df.groupby(["city"]).mean()["driver_count"]
total_urban_drivers = urban_driver_count.sum()

suburban_driver_count = suburban_cities_df.groupby(["city"]).mean()["driver_count"]
total_suburban_drivers = suburban_driver_count.sum()

rural_driver_count = rural_cities_df.groupby(["city"]).mean()["driver_count"]
total_rural_drivers = rural_driver_count.sum()
```

```python
# Total fares
urban_sum_fare = urban_cities_df.groupby(["city"]).sum()["fare"]
total_fare_urban = urban_sum_fare.sum()

suburban_sum_fare = suburban_cities_df.groupby(["city"]).sum()["fare"]
total_fare_suburban = suburban_sum_fare.sum()

rural_sum_fare = rural_cities_df.groupby(["city"]).sum()["fare"]
total_fare_rural = rural_sum_fare.sum()
```

```python
# Average fare per ride
urban_fares = urban_cities_df["fare"]
mean_urban_fares = np.mean(urban_fares)

suburban_fares = suburban_cities_df["fare"]
mean_suburban_fares = np.mean(suburban_fares)

rural_fares = rural_cities_df["fare"]
mean_rural_fares = np.mean(rural_fares)
```

```python
# Average fare per driver
urban_per_driver = total_fare_urban/total_urban_drivers

suburban_per_driver = total_fare_suburban/total_suburban_drivers

rural_per_driver = total_fare_rural/total_rural_drivers
```

```python
# Create empty data frame
df = pd.DataFrame()
```

```python
# Add data to data frame and format
dict = {'Total Rides':[total_rides_rural, total_rides_suburban, total_rides_urban], 'Total Drivers':[total_rural_drivers, total_suburban_drivers, total_urban_drivers], 'Total Fares':[total_fare_rural, total_fare_suburban, total_fare_urban], 'Average Fare per Ride': [mean_rural_fares, mean_suburban_fares, mean_urban_fares], 'Average Fare per Driver':[rural_per_driver, suburban_per_driver, urban_per_driver]}
df = pd.DataFrame(dict, index = ['Rural', 'Suburban', 'Urban'])
df['Total Rides'] = df['Total Rides'].map("{:,}".format)
df['Total Drivers'] = df['Total Drivers'].map("{:,.0f}".format)
df['Total Fares'] = df['Total Fares'].map("${:,.2f}".format)
df['Average Fare per Ride'] = df['Average Fare per Ride'].map("${:,.2f}".format)
df['Average Fare per Driver'] = df['Average Fare per Driver'].map("${:.2f}".format)
df
```

```python
# Rename columns
pyber_data_df = pyber_data_df.rename(columns={'city': 'City', 'date':'Date','fare':'Fare', 'ride_id': 'Ride Id','driver_count': 'No. Drivers', 'type':'City Type'})
```

```python
# Make index Date
pyber_data_df = pyber_data_df.set_index('Date')
```

```python
pyber_data_df
```

```python
# Make new DataFrame for fares
fares_df = pd.DataFrame.copy(pyber_data_df)
```

```python
# Drop other columns that are not City Type, Date, and Fare
fares_df = fares_df.drop(columns=['Ride Id', 'No. Drivers', 'City'])
```

```python
# Set index to datetime data type
fares_df = pd.DataFrame(fares_df, index=pd.DatetimeIndex(fares_df.index))
```

```python
# Check index type
fares_df.info()
```

```python
fares_df
```

```python
# Calculate sum of fares by city type and date
fares_sum_df = pd.DataFrame(fares_df.groupby(["City Type", "Date"]).sum()["Fare"])
```

```python
# Reset Index
fares_sum_df.reset_index
fares_sum_df
```

```python
# Make fares pivot table
fares_pivot = pd.pivot_table(fares_sum_df, index=['Date'], columns = ['City Type'])
fares_pivot
```

```python
# Make new data frame with dates '2019-01-01':'2019-04-28'
fares_pivot_df = pd.DataFrame(fares_pivot)
fares_loc = fares_pivot_df.loc['2019-01-01':'2019-04-28']
```

```python
# Calculate sums for each week
resample_fares_df = pd.DataFrame(fares_loc).resample('W').sum()
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
import matplotlib.style as style
```

```python
# Create the plot
style.use('fivethirtyeight')
resample_fares_df.plot()
plt.xlabel("")
plt.ylabel("Fare($USD)")
plt.title("Total Fare by City Type")
plt.grid()
plt.legend()
plt.savefig("analysis/Fig7.png")
```
